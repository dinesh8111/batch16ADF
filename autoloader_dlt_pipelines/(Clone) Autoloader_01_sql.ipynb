âœ… How to Control Triggers in SQL Auto Loader

Default (Continuous)

If you just CREATE STREAMING TABLE ..., Databricks runs it continuously.

AvailableNow / Trigger Once

When you schedule the query in a Job or Workflow, you choose:

Triggered (batch-like, runs once or on demand)

Continuous (keeps running like PySpark .trigger(processingTime="...")).

So the control moves outside SQL and into the execution environment (Jobs/Pipelines).
