```python
%sql
CREATE OR REPLACE STREAMING TABLE batch16v1_bronze
USING cloud_files
OPTIONS (
  path "/Volumes/sivaadbuc/default/batch16v1",                   -- Source directory
  format "csv",                                                  -- File format
  header "true",                                                 -- CSV header
  delimiter ",",                                                 -- CSV delimiter

  -- Schema Options
  cloudFiles.schemaLocation "/Volumes/sivaadbuc/default/batch16v1/_schema",  -- Schema storage
  cloudFiles.inferColumnTypes "true",                                          -- Infer column types
  cloudFiles.schemaEvolutionMode "addNewColumns",                             -- Schema evolution
  cloudFiles.schemaHints "amount DOUBLE, transaction_date TIMESTAMP",         -- Manual schema hints

  -- File Detection Options
  cloudFiles.useNotifications "true",                                         -- Use Event Grid/Event Hub
  cloudFiles.includeExistingFiles "true",                                     -- Load already existing files
  cloudFiles.maxFilesPerTrigger "1000",                                       -- Limit batch size
  cloudFiles.allowOverwrites "true",                                          -- Allow overwrites

  -- Data Quality Options
  badRecordsPath "/Volumes/sivaadbuc/default/batch16v1/_badrecords",          -- Store bad records
  enforceSchema "true"                                                        -- Strict schema enforcement
);
```
